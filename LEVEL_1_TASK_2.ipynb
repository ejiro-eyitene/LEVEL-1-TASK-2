# TASK 2 - Exploratory Data Analysis(EDA)

```python
##Importing Packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
## Importing DataSet
SPD = pd.read_csv("2) Stock Prices Data Set.csv")
SPD
```

```python
##Calculating Summary Statistics(Mean, Median and Mode)
SPD.describe()
```

## Visualizing Data Distribution

```python
## Ploting Histogram
plt.figure(figsize = (10,6))
plt.hist(SPD.close,bins=50, color = "Lightblue", edgecolor="black")
plt.title("Distribution of Close Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
```

```python
## Ploting BoxPlot
plt.figure(figsize=(8, 6))
sns.boxplot(y=SPD.close, color='lightgreen')
plt.title('Distribution of All Close Prices')
plt.show()
```

```python
## Ploting scatter Plot
plt.figure(figsize=(10,6))
plt.scatter(SPD.open, SPD.close,
           alpha = 0.5,
            color=["skyblue"])
plt.xlabel("Open Price")
plt.ylabel("Close Price")
plt.title("Relationship between Open and Close Prices");
```

## Correlation Between Numerical Features

```python
hd_corr = SPD.corr(numeric_only=True)
ax, fig = plt.subplots(figsize=(10,6))
ax = sns.heatmap(hd_corr, annot=True,
                 linewidths = 0.2,
                 fmt = "0.2f",
                 cmap="twilight_r");
```
